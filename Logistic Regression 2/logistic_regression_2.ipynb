# Logistic Regression

## Part 1 - Data Preprocessing

### Importing the dataset

```python
import pandas as pd
dataset1 = pd.read_excel('train_cleaned.xlsx')
dataset2 = pd.read_excel('test_cleaned.xlsx')
```

```python
dataset1.info()
```

### Getting the inputs and output

```python
dataset1.head(10)
```

```python
dataset2.head(10)
```

```python
# [row,column]
X1 = dataset1.iloc[:,1:8].values
X2 = dataset2.iloc[:,1:8].values
y = dataset1.iloc[:,0].values
```

```python
X1
```

```python
X2
```

```python
y
```

### Creating the Training Set and the Test Set

```python
from sklearn.model_selection import train_test_split
X1_train, X1_test, y_train, y_test = train_test_split(X1, y, test_size = 0.2, random_state = 0)
X2_train, X2_test = train_test_split(X2, test_size = 0.2, random_state = 0)
```

```python
X1_test
```

```python
X2_train
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X1_train = sc.fit_transform(X1_train)
X2_train = sc.fit_transform(X2_train)
```

```python
X1_train
```

```python
X2_train
```

## Part 2 - Building and training the model

### Building the model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state = 0)
```

### Training the model

```python
model.fit(X1_train, y_train)
```

### Inference

Making the predictons of the data points in the test set

```python
y_pred1 = model.predict(sc.transform(X1_test))
y_pred2 = model.predict(sc.transform(X2_test))
```

```python
y_pred1
```

```python
y_pred2
```

```python
y_test
```

## Part 3: Evaluating the model 

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred1)
```
